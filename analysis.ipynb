```python
# Plot settings
import numpy as np

x = np.arange(len(labels)) / 2  # X-axis positions for metrics
bar_width = 0.2  # Width of each bar
offset = bar_width / 2  # Offset for paired bars

plt.figure(figsize=(8,5))  # Adjust figure size

# Plot bars for each category
bar1 = plt.bar(
    x - 0.5 * offset,
    means,
    width=bar_width,
    capsize=5,
    label="500 epochs",
    color="lightblue", 
    edgecolor="black",
)
bar2 = plt.bar(
    x + 1.5 * offset,
    [0.555, 0.6037, 0.506, 0.5795],
    width=bar_width,
    capsize=5,
    label="700 epochs",
    color="lightgreen",
    edgecolor="black",
)

for bar in bar1:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f"{bar.get_height():.2f}",
        ha="center",
        va="bottom",
        fontsize=10,
    )
for bar in bar2:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f"{bar.get_height():.2f}",
        ha="center",
        va="bottom",
        fontsize=10,
    )

# Customize axes and labels
plt.ylim(top=1)
plt.xticks(x + 0.05, labels, rotation=0)
plt.ylabel("DICE Score")
plt.title("")
plt.legend()
plt.tight_layout()

# Display plot
plt.show()
```
